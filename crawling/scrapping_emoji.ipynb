### 2. create emoji_dict

```python
d = {}
ls = response.xpath('body/ul')[:-1]
for l in ls:
    ls2 = l.xpath('./li/div/span')
    al = []; src = []; 
    for j in ls2:
        if 'data-alternative-name' in j.attrib.keys():
            al.append(j.attrib['data-alternative-name'])
        elif 'data-src' in j.attrib.keys():
            src.append(j.attrib['data-src'])
        else:
            print('No data')
            print(j.get())
    d[l.attrib['id']] = {'data-src' : src, 'data-alternative-name' : al}  
d.keys()
```
