# 회귀분석용 가상 데이터 생성
- 기본
## y = b + $w^Tx$ + $\epsilon$


- N개의 y 데이터
## y = b$\cdot$1 + Xw + $\epsilon$
    - y: N Dim vector
    - b: bias, y절편
    - 1; N Dim 1-vector
    - X: N(Sample) X M(Feature) Dim Matrix
    - w: M Dim vector
    - $\epsilon$: N Dim vector, noise or disturbance
    

연습

```python
from sklearn.datasets import make_regression
X, y, w = make_regression(n_samples=100, n_features=1, 
                          bias=10, noise=10, 
                          random_state=0, coef=True)
```

```python
%matplotlib inline
plt.scatter(X, y)
plt.show()
```

함수구현

추가
- `n_informative`: 실제 y에 영향 주는 feature 갯수
    - 정수, 10
- `effective_rank`: feature 중 서로 독립인 갯수(None = 모두 독립
    - 정수, None
- `tail_strength`: 독립변수 간 상관관계 정도(0.5=독립)
(effective_rank=None이면 안됨)
    - 0 - 1, 0.5

### 예제1-1) 두 독립변수 독립, 둘 다 종속 변수에 영향
- `n_informative=2`

```python
X, y, w = make_regression(n_samples=300, n_features=2, noise=10,
                          coef=True, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.show()
```

### 예제1-2) 두 독립변수 독립, 하나만 종속 변수에 영향
- `n_informative=1`

```python
X, y, w = make_regression(n_samples=300, n_features=2, n_informative=1,
                          noise=0, coef=True, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.show()
```

### 예제3) 두 독립변수 상관관계있는 경우(다중공선성)
- `tail_strength=1`

```python
X, y, w = make_regression(n_samples=300, n_features=2, effective_rank=1, noise=0,
                          tail_strength=0.5, coef=True, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.show()
```
