# 확률변수의 독립과 상관

## 1) 독립 - 결합밀도 함수 Joint pdf
- 주변 확률 밀도 함수 Marginal pdf 의 곱
### $$f_{XY}(x, y) = f_{X}(x)f_{Y}(y)$$

## 1.1) 독립 - 반복시행시 pdf
### $$f(x_1,x_2,x_3,\cdots,x_N) = \prod_{i=1}^N f(x_i)$$

## 2) 독립 - 조건부 확률밀도 함수 Conditional pdf
### $$f_{X \mid Y}(x \mid y) = f_{X}(x)$$

### 예) 다변수 이산확률분포 - Joint pmf_independent

```python
%matplotlib inline
#pmf
np.set_printoptions(precision=2)
pmf1 = np.array([[1, 2,  4, 2, 1],
                 [2, 4,  8, 4, 2],
                 [4, 8, 16, 8, 4],
                 [2, 4,  8, 4, 2],
                 [1, 2,  4, 2, 1]])
pmf1 = pmf1/pmf1.sum()

#marginal pmf
pmf1_marginal_x = pmf1.sum(axis=0)
pmf1_marginal_y = pmf1.sum(axis=1)
pmf1x = pmf1_marginal_x * pmf1_marginal_y[:, np.newaxis]

#heatmap
plt.subplot(121)
sns.heatmap(pmf1, cmap=mpl.cm.bone_r, annot=True, square=True, linewidths=0.5, linecolor='k', cbar=False, xticklabels=pmf1_marginal_x, yticklabels=pmf1_marginal_y)
plt.title("Joint pmf")

plt.subplot(122)
sns.heatmap(pmf1x, cmap=mpl.cm.bone_r, annot=True, square=True, linewidths=0.5, linecolor='k', cbar=False, xticklabels=pmf1_marginal_x, yticklabels=pmf1_marginal_y)
plt.title("Joint pmf_independent")

plt.tight_layout()
plt.show()
```

```python
#Conditional pmf_indepent
cond_y0 = pmf1[0, :]/pmf1.sum(axis=1)[0]
cond_y1 = pmf1[1, :]/pmf1.sum(axis=1)[1]
cond_y2 = pmf1[2, :]/pmf1.sum(axis=1)[2]

plt.subplot(131)
plt.bar(np.arange(5), cond_y0)
plt.ylabel('P(x|y=A)', rotation=0, labelpad=30)

plt.subplot(132)
plt.bar(np.arange(5), cond_y1)
plt.ylabel('P(x|y=B)', rotation=0, labelpad=30)

plt.subplot(133)
plt.bar(np.arange(5), cond_y2)
plt.ylabel('P(x|y=C)', rotation=0, labelpad=30)
plt.tight_layout()
plt.show()
```

### 예)Joint pmf_dependent

```python
np.set_printoptions(precision=2)
pmf2 = np.array([[0, 0,  0, 5, 5],
                 [0, 5,  5, 5, 5],
                 [0, 5, 30, 5, 0],
                 [5, 5,  5, 5, 0],
                 [5, 5,  0, 0, 0]])
pmf2 = pmf2 / pmf2.sum()

pmf2_marginal_x = pmf2.sum(axis=0)
pmf2_marginal_y = pmf2.sum(axis=1)
pmf2x = pmf2_marginal_x * pmf2_marginal_y[:, np.newaxis]

plt.subplot(121)
sns.heatmap(pmf2, cmap=mpl.cm.bone_r, annot=True, square=True, linewidth=1, linecolor="k",
            cbar=False, xticklabels=pmf1_marginal_x, yticklabels=pmf1_marginal_y)
plt.title('joint pmf')

plt.subplot(122)
sns.heatmap(pmf2x, cmap=mpl.cm.bone_r, annot=True, square=True, linewidth=1, linecolor="k",
            cbar=False, xticklabels=pmf1_marginal_x, yticklabels=pmf1_marginal_y)
plt.title('joint pmf_dependent')

plt.tight_layout()
plt.show()
```

```python
#Conditional pmf_dependentt
cond_y0 = pmf2[0, :]/pmf2.sum(axis=1)[0]
cond_y1 = pmf2[1, :]/pmf2.sum(axis=1)[1]
cond_y2 = pmf2[2, :]/pmf2.sum(axis=1)[2]

plt.subplot(131).xaxis.set_ticklabels([])
plt.bar(np.arange(5), cond_y0)
plt.ylabel('P(x|y=A)', rotation=0, labelpad=30)

plt.subplot(132).xaxis.set_ticklabels([])
plt.bar(np.arange(5), cond_y1)
plt.ylabel('P(x|y=B)', rotation=0, labelpad=30)

plt.subplot(133).xaxis.set_ticklabels([])
plt.bar(np.arange(5), cond_y2)
plt.ylabel('P(x|y=C)', rotation=0, labelpad=30)
plt.tight_layout()
plt.show()
```

# 공분산과 상관계수

## 1.1)표본 공분산(sample covariance)
- 다변수 확률변수
- 표본평균 $\bar x, \bar y$로 부터 얼마나 멀리있는 정도
- Data 분포의 크기와 방향성
  - 1,3사분면/2,4분면에 밀집했는지 확인 가능
  - x, y 부호같은지 다른지 확인 가능
### $$s_{xy} = \dfrac{1}{N}\sum_{i=1}^{N} (x_i - \bar{x})(y_i-\bar{y})$$

## 1.2)표본 상관계수(sample correlation coefficient)
- 방향성 파악: sample covariance을 각 표본 표준편차값으로 나누어 정규화
- **피어슨 상관계수(Pearson) 상관계수**
### $$r_{xy} = \dfrac{s_{xy}}{\sqrt{s^2_{x} \cdot s^2_{y}}}$$

## 2.1)확률변수 공분산
### $$Cov[X,Y] = E[(X-E[X])(Y-E[Y])]$$

## 2.2)확률변수 상관계수
- $\rho$ : 1, 0, -1  = 완전선형 상관관계, 무상관, 완전선형 반상관관계
### $$\rho[X,Y] = \dfrac{\text{Cov}[X,Y]}{\sqrt{\text{Var}[X] \cdot \text{Var}[Y]}}$$

### 2.2.1) $\rho$의 절대값에 따라 납작, 뚱뚱 차이

```python
#질문:
#x, y = np.random.multivariate_normal([0,0], [[1,r],[r,1]], 1000).T

np.random.seed(0)
corrs = [1, 0.7, 0.3, 0, -0.3, -0.7, -1]
plt.figure(figsize=(len(corrs), 1))
for i, r in enumerate(corrs):
    x, y = np.random.multivariate_normal([0,0], [[1,r],[r,1]], 1000).T
    plt.subplot(1, len(corrs), i + 1)
    plt.plot(x, y, 'bs', ms=0.2)
    plt.axis('equal')
    plt.xticks([])
    plt.yticks([])
    plt.title(r)
plt.show()
```

### 2.2.2) $\rho$와 scater plot기울기는 무관 = 기울기 달라도 $\rho$값은 같음: scaling에 불과함

```python
np.random.seed(1)
slope = [1, 0.7, 0.3, 0, -0.3, -0.7, -1]
plt.figure(figsize=(len(slope), 1))
for i, s in enumerate(slope):
    plt.subplot(1, len(slope), i + 1)
    x, y = np.random.multivariate_normal([0, 0], [[1, 1], [1, 1]], 100).T
    y2 = s * y
    plt.plot(x, y2, 'bo', ms=1)
    plt.axis('equal')
    plt.xticks([])
    plt.yticks([])
    if s > 0:
        plt.title(1)
    if s < 0:
        plt.title(-1)
plt.show()
```

## 3.1)다변수 확률변수의 표본sample 공분산

- N(sample) x M(feature)
- 확률변수 $x_1, x_2, \cdots, x_M$
- 각 열은 확률변수의 표본
$$X = 
\begin{bmatrix}
x_{1,1} & x_{1,2} & \cdots & x_{1,M} \\
x_{2,1} & x_{2,2} & \cdots & x_{2,M} \\
\vdots  &  \vdots & \ddots & \vdots  \\
x_{1,N} & x_{N,2} & \cdots & x_{N,M} \\
\end{bmatrix}$$

- 확률변수$\bar x_{j}$의 기댓값
$$E[x_j] = \bar{x}_j = \dfrac{1}{N} \sum_{i=1}^N x_{ij}$$

- 기댓값 벡터(열로 변환)
$$\bar{x} =
\begin{bmatrix}
\bar{x}_1\\
\bar{x}_2\\
\vdots\\
\bar{x}_M\\
\end{bmatrix}
$$

- 확률변수 $x_j$의 분산
$$Var[x_j] = s_j^2 = \dfrac{1}{N} \sum_{i=1}^N(x_{ij} - \bar{x})^2$$

- 확률변수 $x_j, x_k$의 공분산(행렬 X는 확률변수 표본)
$$\text{Var}[x_j,x_k] = s_{jk} = \dfrac{1}{N} \sum_{i=1}^N (x_{ij} - \bar {x}_j)(x_{ik} - \bar{x}_k)$$

- 표본공분산 행렬(Sample Covariance Matrix) 
    - cf. $X_0$은 평균제거 행렬
    
$$S = 
\begin{bmatrix}
\begin{eqnarray}
s_{x1}^2 \;\;& s_{x_1x_2} \;\;& \cdots \;\;& s_{x_1x_M}\\
s_{x_2x_1} \;\;& s_{x_2}^2 & \cdots & s_{x_2x_M}\\
\vdots \;\;& \vdots \;\;& \ddots &\;\; \vdots \\
s_{x_Mx_1} \;\;& s_{x_Mx_2} \;\;& \cdots \;\;& s_{x_M}^2\\
\end{eqnarray}
\end{bmatrix}
$$

$$
S = \dfrac{1}{N}X_0^TX_0
$$
-$X_0$ 구하는 법
$$
X_0 = X - \dfrac{1}{N}1_M1_N^TX
$$

## 3.2) 다변수 확률변수의 공분산
$$\begin{eqnarray}
\Sigma
&=& \text{Cov}[X] \\
&=& \text{E} \left[ (X - \text{E}[X])(X-\text{E}[X])^T\right]
\\
&=&
\begin{bmatrix}
\sigma_{x_1}^2 & \sigma_{x_1x_2} & \cdots & \sigma_{x_1x_M} \\
\sigma_{x_2x_1} & \sigma_{x_2}^2 & \cdots & \sigma_{x_2x_M} \\
\vdots & \vdots & \ddots & \vdots \\
\sigma_{x_Mx_1} & \sigma_{x_Mx_2} & \cdots & \sigma_{x_M}^2\\
\end{bmatrix}
\end{eqnarray}
$$
