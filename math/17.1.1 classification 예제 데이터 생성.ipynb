보충) plt.subplot 연습

```python
fig = plt.figure(figsize=(6, 6))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0)

columns, rows = 4, 5
h, w = 10, 10

for i in range(1, columns*rows +1):
    img = np.random.randint(400, size=(h, w))
    ax = fig.add_subplot(rows, columns, i)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    plt.imshow(img)
plt.tight_layout()
plt.show()
```

```python
N, M = 2, 5
fig = plt.figure(figsize=(8, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
klist = np.random.choice(range(len(olivetti.data)), N * M)
for i in range(N):
    for j in range(M):
        k = klist[i*M + j]
        ax = fig.add_subplot(N, M, i*M +j+1)
        ax.imshow(olivetti.images[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(olivetti.target[k])
plt.tight_layout()
plt.show()
```
